```python
# Import library
import numpy as np  # Digunakan untuk melakukan perhitungan numerik pada array multidimensi atau matriks
import imageio  # Digunakan untuk membaca dan menulis berbagai jenis file gambar
import matplotlib.pyplot as plt  # Digunakan untuk membuat visualisasi data seperti grafik, plot, histogram, dsb
```

```python
img = imageio.imread("gajah.jpg") #Membaca file gambar 
```

```python
img_height = img.shape[0]  # Memperoleh nilai tinggi gambar dalam piksel
img_width = img.shape[1]  # Memperoleh nilai lebar gambar dalam piksel
img_channel = img.shape[2]  # Memperoleh jumlah channel warna gambar (misalnya, 3 channel untuk gambar RGB)
```

```python
img_grayscale = np.zeros(img.shape, dtype=np.uint8)  # Membuat array kosong berukuran sama dengan gambar awal untuk menyimpan gambar grayscale

for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]  # Mengambil nilai komponen merah dari pixel pada posisi (x,y)
        green = img[y][x][1]  # Mengambil nilai komponen hijau dari pixel pada posisi (x,y)
        blue = img[y][x][2]  # Mengambil nilai komponen biru dari pixel pada posisi (x,y)
        gray = (int(red) + int(green) + int(blue)) / 3  # Menghitung nilai rata-rata dari ketiga komponen warna pada pixel 
        img_grayscale[y][x] = (gray, gray, gray)  # Menyimpan nilai rata-rata tadi sebagai nilai komponen merah, hijau, dan biru pada pixel 

plt.imshow(img_grayscale)  # Menampilkan gambar grayscale
plt.title("Grayscale")  # Menambahkan judul pada gambar
plt.show()  # Menampilkan gambar 

```

```python
hg = np.zeros((256))  # Membuat variabel untuk menyimpan data gambar
```

```python
for x in range(0, 256): # Mendefinisikan loop yang akan berjalan dari 0 sampai 255
    hg[x] = 0 # Mengatur nilai awal histogram untuk setiap nilai pixel menjadi 0
```

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        gray = img_grayscale[y][x][0]  # Mengambil nilai piksel grayscale pada posisi (x,y)
        hg[gray] += 1  # Menambahkan nilai 1 pada histogram pada indeks sesuai dengan nilai piksel grayscale tersebut
```

```python
# plt.figure(figsize=(20, 6))
# plt.plot(hg, color="black", linewidth=2.0)
# plt.show()

bins = np.linspace(0, 256, 100)  # Membuat rentang bin untuk histogram dengan 100 bins dari 0 hingga 256
plt.hist(hg, bins, color="black", alpha=0.5)  # Membuat histogram dengan menggunakan nilai histogram dan rentang bin yang telah dibuat
plt.title("Histogram")  # Menambahkan judul pada plot
plt.show()  # Menampilkan plot histogram
```

```python
for y in range(0, img_height):
    for x in range(0, img_width):
        red = img[y][x][0]   # Mendapatkan nilai intensitas merah pada piksel pada posisi (y, x)
        green = img[y][x][1]  # Mendapatkan nilai intensitas hijau pada piksel pada posisi (y, x)
        blue = img[y][x][2]  # Mendapatkan nilai intensitas biru pada piksel pada posisi (y, x)
        hgr[red] += 1   # Menambahkan 1 pada histogram merah pada posisi yang sesuai dengan intensitas merah piksel
        hgg[green] += 1   # Menambahkan 1 pada histogram hijau pada posisi yang sesuai dengan intensitas hijau piksel
        hgb[blue] += 1   # Menambahkan 1 pada histogram biru pada posisi yang sesuai dengan intensitas biru piksel

bins = np.linspace(0, 256, 100)  # Membuat interval bins pada rentang nilai 0 hingga 256 sebanyak 100 interval
plt.hist(hgr, bins, color="red", alpha=0.5)  # Membuat histogram dari data pada histogram merah
plt.title("Histogram Red")  # Memberikan judul histogram
plt.show()  # Menampilkan histogram

plt.hist(hgg, bins, color="green", alpha=0.5)  # Membuat histogram dari data pada histogram hijau
plt.title("Histogram Green")  # Memberikan judul histogram
plt.show()  # Menampilkan histogram

plt.hist(hgb, bins, color="blue", alpha=0.5)  # Membuat histogram dari data pada histogram biru
plt.title("Histogram Blue")  # Memberikan judul histogram
plt.show()  # Menampilkan histogram
```

```python
hgk = np.zeros((256)) # membuat array hgk dengan 256 elemen nol
c = np.zeros((256)) # membuat array c dengan 256 elemen nol

for x in range(0, 256): # loop sebanyak 256 elemen pada array hgk dan c
    hgk[x] = 0 # set nilai elemen ke-x dari array hgk menjadi 0
    c[x] = 0 # set nilai elemen ke-x dari array c menjadi 0

for y in range(0, img_height): # loop sebanyak tinggi gambar (jumlah piksel baris)
    for x in range(0, img_width): # loop sebanyak lebar gambar (jumlah piksel kolom)
        gray = img_grayscale[y][x][0] # nilai piksel grayscale pada baris ke-y, kolom ke-x
        hgk[gray] += 1 # tambahkan 1 ke elemen ke-gray dari array hgk
                
c[0] = hgk[0] # set nilai elemen ke-0 dari array c dengan nilai elemen ke-0 dari array hgk
for x in range(1, 256): # loop sebanyak 255 elemen pada array hgk dan c
     c[x] = c[x-1] + hgk[x] # set nilai elemen ke-x dari array c dengan penjumlahan elemen ke-(x-1) dan elemen ke-x dari array hgk

hmaxk = c[255] # simpan nilai elemen ke-255 dari array c ke dalam variabel hmaxk

for x in range(0, 256): # loop sebanyak 256 elemen pada array c
    c[x] = 190 * c[x] / hmaxk # set nilai elemen ke-x dari array c dengan perhitungan nilai elemen ke-x dikali 190 dan dibagi dengan hmaxk

plt.hist(c, bins, color="black", alpha=0.5) # plot histogram untuk array c dengan warna hitam transparan
plt.title("Histogram Grayscale Kumulatif") # judul plot histogram
plt.show() # tampilkan plot histogram
```

```python
hgh = np.zeros((256)) # mengambil array hgh dengan 256 elemen nol
h = np.zeros((256)) # mengambil array h dengan 256 elemen nol
c = np.zeros((256)) # mengambil array c dengan 256 elemen nol


for x in range(0, 256): # loop sebanyak 256 elemen pada array hgh, h, dan c
    hgh[x] = 0 # set nilai elemen ke-x dari array hgh menjadi 0
    h[x] = 0 # set nilai elemen ke-x dari array h menjadi 0
    c[x] = 0 # set nilai elemen ke-x dari array c menjadi 0

for y in range(0, img_height): # loop sebanyak tinggi gambar (jumlah piksel baris)
    for x in range(0, img_width): # loop sebanyak lebar gambar (jumlah piksel kolom)
        gray = img_grayscale[y][x][0] # nilai piksel grayscale pada baris ke-y, kolom ke-x
        hgh[gray] += 1 # tambahkan 1 ke elemen ke-gray dari array hgh
                
h[0] = hgh[0] # set nilai elemen ke-0 dari array h dengan nilai elemen ke-0 dari array hgh
for x in range(1, 256): # loop sebanyak 255 elemen pada array h dan hgh
     h[x] = h[x-1] + hgh[x] # set nilai elemen ke-x dari array h dengan penjumlahan elemen ke-(x-1) dan elemen ke-x dari array hgh

for x in range(0, 256): # loop sebanyak 256 elemen pada array h
     h[x] = h[x] / img_height / img_width # set nilai elemen ke-x dari array h dengan perhitungan nilai elemen ke-x dibagi dengan hasil kali tinggi dan lebar gambar

for x in range(0, 256): # loop sebanyak 256 elemen pada array hgh
    hgh[x] = 0 # set nilai elemen ke-x dari array hgh menjadi 0
    
for y in range(0, img_height): # loop sebanyak tinggi gambar (jumlah piksel baris)
    for x in range(0, img_width): # loop sebanyak lebar gambar (jumlah piksel kolom)
        gray = img_grayscale[y][x][0] # nilai piksel grayscale pada baris ke-y, kolom ke-x
        gray = h[gray] * 255 # ubah nilai piksel grayscale menjadi nilai baru dengan mengalikan nilai piksel grayscale sebelumnya dengan 255 dan kemudian membaginya dengan nilai elemen ke-nilai piksel grayscale baru dari array h
        hgh[int(gray)] += 1 # tambahkan 1 ke elemen ke-nilai piksel grayscale baru yang dibulatkan menjadi integer dari array hgh

c[0] = hgh[0] # set nilai elemen ke-0 dari array c dengan nilai elemen ke-0 dari array hgh
for x in range(1, 256): # loop sebanyak 255 elemen pada array hgh dan c
     c[x] = c[x-1] + hgh[x] # set nilai elemen ke-x dari array c dengan penjumlahan elemen ke-(x-1) dan elemen ke-x dari array hgh

hmaxk = c[255] # Mencari nilai elemen ke-255 dari array 'c' dan menyimpannya ke dalam variabel 'hmaxk'

for x in range(0, 256): #  Melakukan loop pada range 0 hingga 255
    c[x] = 190 * c[x] / hmaxk # Melakukan normalisasi pada elemen ke-x array c

plt.hist(c, bins, color="black", alpha=0.5) # Membuat histogram dengan data yang telah di-normalisasi dari array c
plt.title("Histogram Grayscale Hequalisasi") # Memberikan judul pada plot histogram
plt.show() # Menampilkan plot histogram pada layar
```
