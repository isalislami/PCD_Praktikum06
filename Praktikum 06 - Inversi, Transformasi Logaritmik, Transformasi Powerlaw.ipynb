```python
# Import library
import numpy as np  # Digunakan untuk melakukan perhitungan numerik pada array multidimensi atau matriks
import imageio  # Digunakan untuk membaca dan menulis berbagai jenis file gambar
import matplotlib.pyplot as plt  # Digunakan untuk membuat visualisasi data seperti grafik, plot, histogram, dsb
```

```python
img = imageio.imread("gajah.jpg") #Membaca file gambar 
```

```python
img_height = img.shape[0]  # Memperoleh nilai tinggi gambar dalam piksel
img_width = img.shape[1]  # Memperoleh nilai lebar gambar dalam piksel
img_channel = img.shape[2]  # Memperoleh jumlah channel warna gambar (misalnya, 3 channel untuk gambar RGB)
```

```python
img_inversi = np.zeros(img.shape, dtype=np.uint8) #Membuat array img_inversi dengan ukuran yang sama dengan img dan tipe data uint8
```

```python
def inversi_grayscale(nilai):
    # Looping setiap piksel dalam gambar
    for y in range(0, img_height):
        for x in range(0, img_width):
            # Mendapatkan nilai intensitas warna RGB pada piksel pada posisi (y, x)
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            # Menghitung nilai rata-rata intensitas warna RGB untuk membuat gambar grayscale
            gray = (int(red) + int(green) + int(blue)) / 3
            # Melakukan operasi inversi pada nilai grayscale
            gray = nilai - gray
            # Menetapkan nilai piksel inversi pada gambar baru
            img_inversi[y][x] = (gray, gray, gray)
```

```python
def inversi_rgb(nilai):
    # Looping setiap piksel pada gambar
    for y in range(0, img_height):
        for x in range(0, img_width):
            # Mendapatkan nilai intensitas merah, hijau, dan biru pada piksel
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]

            # Melakukan inversi pada nilai intensitas merah, hijau, dan biru dengan nilai yang diberikan
            red = nilai - red
            green = nilai - green
            blue = nilai - blue

            # Memasukkan nilai intensitas merah, hijau, dan biru yang sudah di-inversi ke dalam gambar hasil inversi
            img_inversi[y][x] = (red, green, blue)


```

```python
# Melakukan inversi gambar grayscale dengan nilai maksimum 255
inversi_grayscale(255)

# Menampilkan gambar hasil inversi grayscale
plt.imshow(img_inversi)
plt.title("Inversi Grayscale")
plt.show()

# Melakukan inversi gambar RGB dengan nilai maksimum 255
inversi_rgb(255)

# Menampilkan gambar hasil inversi RGB
plt.imshow(img_inversi)
plt.title("Inversi RGB")
plt.show()
```

```python
img_log = np.zeros(img.shape, dtype=np.uint8) # Mengambil variabel img_grayscale dengan nilai zero array
```

```python
def log(c):    # mendefinisikan fungsi dengan nama 'log' dan parameter 'c'
    for y in range(0, img_height):    # melakukan perulangan untuk setiap baris gambar
        for x in range(0, img_width):    # melakukan perulangan untuk setiap piksel pada setiap baris
            red = img[y][x][0]    # mengambil nilai komponen merah piksel pada koordinat (x,y)
            green = img[y][x][1]    # mengambil nilai komponen hijau piksel pada koordinat (x,y)
            blue = img[y][x][2]    # mengambil nilai komponen biru piksel pada koordinat (x,y)
            gray = (int(red) + int(green) + int(blue)) / 3    # menghitung nilai rata-rata intensitas piksel untuk mendapatkan citra grayscale
            gray = int(c * np.log(gray + 1))    # mengaplikasikan fungsi logaritma pada nilai grayscale dengan parameter 'c'
            if gray > 255:    # memastikan bahwa nilai grayscale tidak melebihi 255 (nilai maksimum)
                gray = 255
            if gray < 0:    # memastikan bahwa nilai grayscale tidak kurang dari 0 (nilai minimum)
                gray = 0
            img_log[y][x] = (gray, gray, gray)    # menyimpan nilai grayscale hasil transformasi pada citra yang baru (img_log)
```

```python
img_log = np.zeros(img.shape, dtype=np.uint8) # membuat matriks kosong dengan ukuran yang sama dengan gambar asli
log(30) # melakukan transformasi log pada gambar
plt.imshow(img_log) # menampilkan gambar hasil transformasi menggunakan matplotlib
plt.title("Log") # menambahkan judul pada plot gambar
plt.show() # menampilkan plot gambar
```

```python
img_inlog = np.zeros(img.shape, dtype=np.uint8) #Membuat array kosong dengan ukuran yang sama dengan gambar asli 
```

```python
def inlog(c): # mendefinisikan fungsi inlog dengan satu parameter yaitu c
    for y in range(0, img_height): # loop dari 0 sampai img_height (tinggi gambar)
        for x in range(0, img_width): # loop dari 0 sampai img_width (lebar gambar)
            red = img[y][x][0] # mengambil nilai pixel merah dari gambar
            green = img[y][x][1] # mengambil nilai pixel hijau dari gambar
            blue = img[y][x][2] # mengambil nilai pixel biru dari gambar
            gray = (int(red) + int(green) + int(blue)) / 3 # menghitung nilai grayscale dari pixel tersebut
            gray = int(c * np.log(255 - gray + 1)) # menghitung nilai transformasi log dari grayscale
            if gray > 255: # jika nilai grayscale melebihi 255, maka diatur ke 255
                gray = 255
            if gray < 0: # jika nilai grayscale kurang dari 0, maka diatur ke 0
                gray = 0
            img_inlog[y][x] = (gray, gray, gray) # menetapkan nilai pixel baru ke img_inlog
```

```python
inlog(30)                         # Memanggil fungsi inlog dengan parameter c = 30
plt.imshow(img_inlog)            # Menampilkan gambar hasil inversi dan log dengan menggunakan pyplot
plt.title("Inversi & Log")       # Menambahkan judul pada gambar
plt.show()                        # Menampilkan gambar pada layar
```

```python
img_nthpower = np.zeros(img.shape, dtype=np.uint8) #menyimpan hasil operasi
```

```python
def nthpower(c, y):
    thc = c / 100      # menghitung thc sebagai presentase nilai c
    thy = y / 100      # menghitung thy sebagai presentase nilai y
    for y in range(0, img_height):       # melakukan looping setiap pixel pada sumbu y
        for x in range(0, img_width):    # melakukan looping setiap pixel pada sumbu x
            red = img[y][x][0]      # mengambil nilai komponen merah pada pixel
            green = img[y][x][1]    # mengambil nilai komponen hijau pada pixel
            blue = img[y][x][2]     # mengambil nilai komponen biru pada pixel
            gray = (int(red) + int(green) + int(blue)) / 3     # menghitung nilai rata-rata dari ketiga komponen warna
            gray = int(thc * pow(gray, thy))   # menghitung nilai gray yang baru menggunakan rumus nth power
            if gray > 255:    # memastikan nilai gray tidak melebihi 255
                gray = 255
            if gray < 0:      # memastikan nilai gray tidak kurang dari 0
                gray = 0
            img_nthpower[y][x] = (gray, gray, gray)   # memasukkan nilai gray ke dalam masing-masing komponen warna pixel
```

```python
nthpower(50, 100)    # Memanggil fungsi nthpower dengan parameter c=50 dan y=100
plt.imshow(img_nthpower)    # Menampilkan gambar yang sudah diolah dengan menggunakan fungsi nthpower
plt.title("Nth Power")    # Menambahkan judul pada gambar
plt.show()    # Menampilkan gambar
```

```python
img_nthrootpower = np.zeros(img.shape, dtype=np.uint8) #menyimpan hasil operasi
```

```python
def nthrootpower(c, y):
    thc = c / 100  # Mengkonversi nilai c ke dalam satuan persen
    thy = y / 100  # Mengkonversi nilai y ke dalam satuan persen
    
    # Melakukan iterasi untuk setiap piksel dalam gambar
    for y in range(0, img_height):
        for x in range(0, img_width):
            # Mengambil nilai red, green, blue dari piksel yang sedang diiterasi
            red = img[y][x][0]
            green = img[y][x][1]
            blue = img[y][x][2]
            # Menghitung nilai gray berdasarkan rata-rata dari red, green, blue
            gray = (int(red) + int(green) + int(blue)) / 3
            # Melakukan perhitungan nilai gray baru menggunakan rumus nth root power
            gray = int(thc * pow(gray, 1./thy))
            # Membatasi nilai gray antara 0-255
            if gray > 255:
                gray = 255
            if gray < 0:
                gray = 0
            # Mengisi piksel dalam gambar hasil dengan nilai gray yang baru
            img_nthpower[y][x] = (gray, gray, gray)
```

```python
#Menampilkan gambar hasil transformasi dengan menggunakan pyplot
nthrootpower(50, 100)
plt.imshow(img_nthrootpower)
plt.title("Nth Root Power")
plt.show()
```
